```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import math
filepath1 = r"C:\Users\lenovo\Desktop\SPX.csv"
df1 = pd.read_csv(filepath1)  
df1['time'] = pd.to_datetime(df1['time'])
df=df1[(df1['time'] >=pd.to_datetime('09:30:00')) & (df1['time'] <= pd.to_datetime('15:00:00'))]
df.head()
groups = df.groupby(df.date)
RV_5min = []
#RV_5min    
for i,group in groups:
    #
    a=[]
    for i in range(0,len(group),5):#5
        a.append(i) 
    group=group.iloc[a] 
    group["close_lag"] = group["close"].shift(1)
    group["ln_clo"] = group["close"].apply(np.log) 
    group["ln_clo_lag"] = group["close_lag"].apply(np.log) 
    group["r"]=group["ln_clo"]-group["ln_clo_lag"]
    group["r^2"]=group["r"]*group["r"]
    rv = group['r^2'].sum()  
    RV_5min.append(rv)  
date=df[["date"]]#
date.drop_duplicates(subset=None, keep='first', inplace=True)# 
date=date.reset_index(drop=True)#

#listdataframe  

RV_5 = pd.DataFrame([RV_5min],index=["RV_5min"])


#

RV_5_2 = RV_5.T


#RV
RV_SZ=pd.concat([date,RV_5_2],axis=1)   

```

```python
writer = pd.ExcelWriter(r"C:\Users\lenovo\Desktop\RV.xlsx")
RV_SZ.to_excel(writer, "sheet1")
writer.save()
```
